#Human Biomass over time

```
to_carbon = 0.15
human_carbon_mass = 5e4*to_carbon
```

```
humans = pd.read_csv('/content/WorldPop.csv')
humans = humans[['Year', 'World Population']]
humans = humans.set_index('Year')
humans['Human BioMass'] = humans['World Population'] *human_carbon_mass*1e-15
humans['% Total Mammal Biomass (Gt C)'] = humans['Human BioMass']/(550 * 3 * 1e-6)
humans.to_csv(r'/content/human_bm.csv')
```

```
birds_bm = 0.007 #Gt C
```

```
area = ['Africa', 'Americas', 'Asia', 'Eastern Europe', 'Northern America', 'Oceania', 'Southern Asia', 'Western Europe']
```

```
birds = pd.read_csv('/content/birds.csv')
birds = birds[birds['Area'].isin(area)]
birds = birds[['Area', 'Item', 'Year', 'Value', 'Unit']]
set(birds.Item)
```

#Chicken BioMass

```
chicken = birds[birds['Item'] == 'Chickens']
chicken = chicken.pivot_table(values = "Value", index = "Area", columns="Year")
chicken.loc['Americas'] -= chicken.loc['Northern America']
chicken.loc['Asia'] -= chicken.loc['Southern Asia']
chicken = chicken.rename(index={'Americas': 'Latin America', 'Southern Asia': 'Indian Subcontinent'})
chicken = chicken.sort_index()
#chicken = chicken.sum()
#chicken *= 1000
#chicken.to_csv(r'/content/chicken.csv')
chicken
```

```
chicken_layer = pd.read_csv('/content/FAOSTAT_egg_chicken.csv')
chicken_layer = chicken_layer[(chicken_layer['Element'] == 'Laying') & (chicken_layer['Area'].isin(area))]
chicken_layer = chicken_layer[['Area', 'Item', 'Year', 'Value', 'Unit']]
chicken_layer = chicken_layer.pivot_table(values = "Value", index = "Area", columns="Year")
chicken_layer.loc['Americas'] -= chicken_layer.loc['Northern America']
chicken_layer.loc['Asia'] -= chicken_layer.loc['Southern Asia']
chicken_layer = chicken_layer.rename(index={'Americas': 'Latin America', 'Southern Asia': 'Indian Subcontinent'})
chicken_layer = chicken_layer.sort_index()
```

```
chicken -= chicken_layer
chicken *= 0.9
chicken_layer *= 1.8
chicken += chicken_layer
chicken = chicken.sum()*converter
#
#chicken
```

```
chicken = chicken.to_frame()
chicken['% Total Animal Biomass (Gt C)'] = (chicken.iloc[:,0])/(550 * 3 * 1e-6)
chicken = chicken.rename(columns={0: 'Total Biomass (Gt C)'})
chicken['% Total of birds'] = (chicken['Total Biomass (Gt C)'] / birds_bm)*1e5
chicken.to_csv(r'/content/chicken.csv')
```

#Duck BioMass

```
ducks = birds[birds['Item'] == 'Ducks']
ducks = ducks.pivot_table(values = "Value", index = "Area", columns="Year")
ducks.loc['Americas'] -= ducks.loc['Northern America']
ducks.loc['Asia'] -= ducks.loc['Southern Asia']
ducks = ducks.sort_index()
ducks *= 2.7
ducks = ducks.sum()*converter
ducks = ducks.to_frame()
ducks['% Total Animal Biomass (Gt C)'] = (ducks.iloc[:,0])/(550 * 3 * 1e-6)
ducks = ducks.rename(columns={0: 'Total Biomass (Gt C)'})
ducks['% Total of birds'] = (ducks['Total Biomass (Gt C)'] / birds_bm)*1e5
#ducks
ducks.to_csv(r'/content/ducks.csv')
```
